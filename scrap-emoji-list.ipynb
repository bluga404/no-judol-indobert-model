## `Scrape Emoji List Data`

```python
import requests
from bs4 import BeautifulSoup
import csv

URL = "https://unicode.org/emoji/charts/full-emoji-list.html"

def scrape_emoji_to_csv(output_file="data/emoji_list.csv"):
    response = requests.get(URL)
    soup = BeautifulSoup(response.text, "html.parser")

    rows = soup.find_all("tr")

    data = []

    for row in rows:
        cols = row.find_all("td")

        if len(cols) < 9:
            continue

        try:
            number = cols[0].text.strip()
            code = cols[1].text.strip()
            emoji = cols[2].text.strip()
            name = cols[8].text.strip()

            data.append([number, code, emoji, name])

        except Exception:
            continue

    # save to CSV
    with open(output_file, "w", newline="", encoding="utf-8") as f:
        writer = csv.writer(f)
        writer.writerow(["No", "Code", "Emoji", "Name"])
        writer.writerows(data)

    print(f"Saved {len(data)} emojis to {output_file}")


# run
scrape_emoji_to_csv()
```

```python
INPUT_FILE = "data/emoji_list.csv"
OUTPUT_FILE = "data/emoji_list_clean.csv"

# keyword yang mau dihapus
REMOVE_KEYWORDS = [
    "aries", "taurus", "gemini", "cancer", "leo", "virgo",
    "libra", "scorpio", "sagittarius", "capricorn", "aquarius", "pisces", "ophiuchus",
    "currency", "dollar",
    "copyright", "registered", "trade mark",
    "keycap",
    "input",
    "button",
    "information",
    "circled"
]

def should_remove(name: str) -> bool:
    name = name.lower()
    return any(keyword in name for keyword in REMOVE_KEYWORDS)


def clean_csv():
    with open(INPUT_FILE, "r", encoding="utf-8") as infile, \
         open(OUTPUT_FILE, "w", newline="", encoding="utf-8") as outfile:

        reader = csv.reader(infile)
        writer = csv.writer(outfile)

        header = next(reader)
        writer.writerow(header)

        for row in reader:
            name = row[3]

            if not should_remove(name):
                writer.writerow(row)

    print("Cleaned CSV saved!")

clean_csv()
```
